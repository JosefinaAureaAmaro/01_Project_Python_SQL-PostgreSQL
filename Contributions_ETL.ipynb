### Transform

I'm going to transform in a loop to make minor adjustments efficent throughout dataframes

My objective is to create a consolidated dataframe of all csv's with dropped columns per candidate. 
 1. Create a list of the dataframes to loop - List is to print the names of dataframe being processed
 2. Create a dictionary of the dataframes to loop - Dictionary is to keep the type of the value 

```python
dict_of_df = {0: hillary_contr_0_2015,
                      1: hillary_contr_1_2015,
                      2: hillary_contr_0_2016_q1, 
                      3: hillary_contr_1_2016_q1, 
                      4: hillary_contr_2_2016_q2, 
                      5: hillary_contr_3_2016_q2, 
                      6: hillary_contr_4_2016_q3, 
                      7: hillary_contr_5_2016_q3, 
                      8: hillary_contr_6_2016_q3, 
                      9: hillary_contr_7_2016_q3, 
                      10: hillary_contr_8_2016_q4,
                      11: hillary_contr_9_2016_q4,
                      12: hillary_contr_10_2016_q4, 
                      13: hillary_contr_11_2016_q4,
                      14: dt_contr_1,
                      15: dt_contr_2}
                     


list_of_df = ['hillary_contr_0_2015',
                      'hillary_contr_1_2015',
                      'hillary_contr_0_2016_q1', 
                      'hillary_contr_1_2016_q1', 
                      'hillary_contr_2_2016_q2', 
                      'hillary_contr_3_2016_q2', 
                      'hillary_contr_4_2016_q3', 
                      'hillary_contr_5_2016_q3', 
                      'hillary_contr_6_2016_q3', 
                      'hillary_contr_7_2016_q3', 
                      'hillary_contr_8_2016_q4',
                      'hillary_contr_9_2016_q4',
                      'hillary_contr_10_2016_q4', 
                      'hillary_contr_11_2016_q4',
                      'dt_contr_1',
                      'dt_contr_2']
```

```python
## Create df of all 2015 - 2016 contr file - Hillary Clinton & Donald Trump

i = 0 
contributors = []
while i <= 15:
    for i in dict_of_df: ## i is acting as key value
        df_name = list_of_df[i] ## records which database we are iterating through   
        contributors.append(dict_of_df[i]) ## append dataframe to list
        i= i+1 ## loop through next dataframe 
        print(f"Completed Concat of {df_name}") #state which dataframes are complete
    contributor_df = pd.concat(contributors,ignore_index=True) # combine to make 1 long df 
```

```python
contributor_df.tail()
```

```python
#to drop columns in list
cleaned_contributor_df = contributor_df.drop(columns= hc_drop_cols_list )

#to drop empty columns
Contributor_df = cleaned_contributor_df.dropna(axis='columns', how='all')  ## drop empty cols
```
