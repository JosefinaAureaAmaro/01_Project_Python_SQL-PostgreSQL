### Transform

I'm going to transform in a loop to make minor adjustments efficent throughout dataframes

My objective is to create a consolidated dataframe of all csv's with dropped columns per candidate. 
 1. Create a list of the dataframes to loop - List is to print the names of dataframe being processed
 2. Create a dictionary of the dataframes to loop - Dictionary is to keep the type of the value 

```python
dict_of_hc_df = {0 : hillary_contr_1, 1 :hillary_contr_2, 2 : hillary_contr_3, 3: hillary_contr_4, 4 : hillary_contr_5, 5 : hillary_contr_6, 6: hillary_contr_7, 7: hillary_contr_8}
list_of_df = {0 : hillary_contr_1, 1 :hillary_contr_2, 2 : hillary_contr_3, 3: hillary_contr_4, 4 : hillary_contr_5, 5 : hillary_contr_6, 6: hillary_contr_7,'hillary_contr_8']
```

```python
## Create new df of country codes 

i = 0 
code_list = []
while i <= 5:
    for i in dict_of_df: ## i is acting as key value
        df_name = list_of_df[i] ## records which database we are iterating through 
        cou = dict_of_df[i][['Country','COU']].groupby('Country')['COU'].unique().reset_index(name="COU").sort_values(by='Country')
        """
        -iterate through dataframes
        -selects two columns for country code
        -print only unique list of COU, 
        -set series values name to 'COU' to make dataframe
        - sort countries name in alphabetical order
        
        """        
        code_list.append(cou) ## append dataframe to list
        i= i+1 ## loop through next dataframe 
        print(f"Completed Concat of {df_name}") #state which dataframes are complete
    code_df = pd.concat(code_list,ignore_index=True) # combine as 1 long list of unique countries
```
